```python
import librosa
import soundfile as sf
import speech_recognition as sr
import matplotlib.pyplot as plt
import numpy as np
import librosa.display
import os
```

```python
workdir = 'SpeechRec/'
os.chdir(workdir)
```

```python
def plot_signal(y, sample_rate, figsize=(14, 8)):
    fig = plt.figure(figsize=figsize)
    ax1 = fig.add_subplot(211)
    ax1.set_xlabel('time, sec')
    ax1.set_ylabel('Amplitude')
    audio_len = y.shape[0] / sample_rate
    x = np.linspace(0, audio_len, len(y))
    ax1.plot(x, y)
    plt.show()
    
def ogg_to_wav(input_filename, output_filename, resample=None):
    y, sm = librosa.load(input_filename)
    if resample is not None:
        y = librosa.resample(y, orig_sr=sm, target_sr=resample)
        sm = resample
    sf.write(output_filename, y, sm, 'PCM_24')
    print(f'File {input_filename} was converted.\n New wav file: {output_filename}')
    return y, sm
    
def speech_to_text(audio_file, audio_len, duration=30):
    r = sr.Recognizer()
    myaudio = sr.AudioFile(audio_file)
    message = ''
    offset = 0
    duration = duration
    while offset < audio_len:
        with myaudio as source:
            audio = r.record(source, offset=offset, duration=duration)
            message += r.recognize_google(audio) + '\n'
        offset += duration
    return message
```

# Raw wav

```python
y, sm = ogg_to_wav(input_filename='audio_2022-02-14_13-51-19.ogg', 
                   output_filename='audio_2022-02-14_13-51-19.wav')
audio_len = y.shape[0] / sm
print(f'Total time: {audio_len} sec')
plot_signal(y, sm)
message_raw = speech_to_text('audio_2022-02-14_13-51-19.ogg.wav', audio_len)
print(message_raw)
```

# Resample 8000

```python
y, sm = ogg_to_wav(input_filename='audio_2022-02-14_13-51-19.ogg', 
                   output_filename='audio_2022-02-14_13-51-19.wav',
                   resample=8000)
audio_len = y.shape[0] / sm
print(f'Total time: {audio_len} sec')
plot_signal(y, sm)
message_8000 = speech_to_text('audio_2022-02-14_13-51-19.ogg.wav', audio_len)
print(message_8000)
```

# Speech to text pipeline

```python
input_filename = 'audio_2022-02-14_18-19-50.ogg'
```

```python
def convert_to_text_from_ogg(input_filename):
    output_filename = input_filename.split('.')[0] + '.wav'
    y, sm = ogg_to_wav(input_filename, 
                       output_filename,
                       16000)
    audio_len = y.shape[0] / sm
    print(f'Total time: {audio_len} sec')
    plot_signal(y, sm)
    message = speech_to_text(output_filename, audio_len)
    return message
```

```python
message = convert_to_text_from_ogg(input_filename)
print(message)
```
